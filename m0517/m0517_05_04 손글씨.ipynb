```python
def test(arr):
    output=[]
    for i in arr:
        output.append(float(i)/256)
    return output   # 0~1 사이의 값
```

```python
# train,test 데이터를 생성
train_input = list(map(test,train_csv.iloc[:,1:].values))   # [[],[],[],[].....[],[]]
test_input = list(map(test,tk_csv.iloc[:,1:].values))
```

```python
# target result 데이터를 생성
train_target = train_csv[0].values      #1001개
test_target = tk_csv[0].values          #501개
```

```python
# 데이터 학습
clf = svm.SVC()
clf.fit(train_input,train_target)

# 데이터 예측
clf.predict(test_input)

```

```python
# 정답률
clf.score(test_input,test_target)
```
