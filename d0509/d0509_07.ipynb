#### 데이터 수정: columns 수정

```python
#pandas: 데이터 컬럼 전체적으로 수정하는 형태
df['학교'] = '디지털고'
df
```

```python
# 학교 컬럼 중 구로고를 디지털고로 변경
df['학교'].replace({'구로고':'디지털고'},inplace=True)
df 
```

```python
filt = df['SW특기'].str.lower().str.contains('java',na=False)
df[filt]
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df.fillna("미선택",inplace=True)
df
```

```python
# 문구 추가 수정.
df['학교'] = df['학교']+'등학교'
df
```

```python
# 과학 점수를 +5점을 더해서 출력
df['과학'] = df['과학'] + 5
df
```

```python
# 국어 점수 +10점
df['국어'] = df['국어'] +10
```
