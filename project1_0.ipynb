```python
# data 경로 설정
data_dir = 'C:/workspace/medical1/z20_data/osteoarthritis/train/0'
categories = ['train','test','val'] # 전처리될 데이터셋을 훈련용, 테스트용, 검증용으로 구분

# 샘플 이미지 시각화
for i in range(9):
    img_name = os.listdir(data_dir)[i]
    img_path = os.path.join(data_dir, img_name)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

for i in range(9):
    plt.figure(figsize=(12,12))
    plt.subplot(3,3,i+1)
    plt.imshow(img, cmap='gray')
    plt.title(i)
    plt.show()
```
