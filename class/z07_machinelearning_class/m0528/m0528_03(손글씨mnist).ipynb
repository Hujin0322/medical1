```python
import numpy as np
import pandas as pd
from tensorflow import keras
```

```python
# keras 라이브러리에 손글씨 데이터 내장되어 있음.
(train_input, train_target),(test_input, test_target) = keras.datasets.mnist.load_data()
train_input.shape
```

```python
train_data=pd.read_csv('../z20_data/train.csv',header=None)
train_data.head(2)
test_data = pd.read_csv('../z20_data/t10k.csv', header=None)
test_data.head(2)
```

```python
# data, target 분리
train_input = train_data.iloc[:,1:] # 1열 삭제
train_target = train_data[0]
test_input = test_data.iloc[:,1:] # 1열 삭제
test_target = test_data[0]
```

```python
train_input
```

```python
# 0-1까지의 값으로 변경
train_input = train_input/225
test_input = test_input/225
```

```python
# 딥러닝 - 인공신경망 모델 
from sklearn.model_selection import train_test_split

# 딥러닝 - 훈련 세트, 검증 세트, 테스트 세트(train, val, test)
train_input, val_input, train_target, val_target = train_test_split(
    train_input, train_target, test_size=0.2, random_state=42
)
```

```python
dense = keras.layers.Dense(10, activation='softmax',input_shape=(784,))
model = keras.Sequential([dense])
model.compile(loss='sparse_categorical_crossentropy',metrics=['accuracy'])
# default: batch_size=32(64,128)
model.fit(train_input,train_target, epochs=5)
model.evaluate(val_input,val_target)
```
