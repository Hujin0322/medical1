```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 설정
matplotlib.rcParams['font.size'] = 10 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df1 = pd.read_excel('../z20_data/xlsx/20230104_연령별인구현황.xlsx',skiprows=3, index_col='행정기관',nrows=1,usecols='B,E:Y')
df1.loc['전국  '] = (df1.loc['전국  '].str.replace(',','').astype(int))//1000
y1 = df1.loc['전국  '].values
df1
```

```python
df2 = pd.read_excel('../z20_data/xlsx/20230104_연령별인구현황.xlsx',skiprows=3, index_col='행정기관',nrows=1,usecols='B,AB:AV')
df2.loc['전국  '] = (df2.loc['전국  '].str.replace(',','').astype(int))//1000
y2 = df2.loc['전국  '].values
df2
```

```python
# 2023년 01월 연령별 인구 수 남, 여 나누어 가로 막대그래프로 출력
x = df1.loc['전국  '].index
plt.barh(x,y1)
plt.barh(x,-y2)
plt.title('2023년 01월 대한민국 인구 피라미드')
```

```python
# 남자 1-4월 각 연령별 합
df = pd.read_excel('../z20_data/xlsx/20230104_연령별인구현황.xlsx',skiprows=3, index_col='행정기관',nrows=1)
df.loc['전국  '] = df.loc['전국  '].astype(str).str.replace(',','').astype(int)
```

```python
title = ['0~4세', '5~9세', '10~14세', '15~19세','20~24세', '25~29세', '30~34세','35~39세', '40~44세', '45~49세', '50~54세',
       '55~59세', '60~64세', '65~69세', '70~74세', '75~79세', '80~84세','85~89세','90~94세', '95~99세', '100세 이상']
y_m = []
y_w = []

for i in title:
    y_m.append(df.loc['전국  '][i] + df.loc['전국  '][i+'.2'] + df.loc['전국  '][i+'.4'] + df.loc['전국  '][i+'.6'])
    y_w.append(-(df.loc['전국  '][i+'.1'] + df.loc['전국  '][i+'.3'] + df.loc['전국  '][i+'.5'] + df.loc['전국  '][i+'.7']))
```

```python
x = title
plt.barh(x,y_m)
plt.barh(x,y_w)
```
