```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 설정
matplotlib.rcParams['font.size'] = 10 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
# 개봉연도별 평점 선 그래프 출력
# 선 = red, marker 표시
# x축 눈금 표시- 2005,2010,2015,2020/ y축 눈금 범위 - 7, 10
```

```python
filt = df.groupby('개봉 연도')['평점'].mean()
filt.index
```

```python
x = filt.index
y = filt.values
plt.plot(x,y,color='red',marker='o',ls='--')
plt.xticks(x)
for i, txt in enumerate(y):
    plt.text(x[i],y[i]+0.1,txt,ha='center',color='blue')
plt.ylim(7,10)
plt.xlabel('개봉 연도')
plt.ylabel('평점')
plt.grid(linestyle='--',alpha=0.5)
```
