```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 합계 컬럼 생성/ 3학년 중 합계 성적이 가장 낮은 학생 출력.
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
```

```python
# 3학년 중에 합계가 가장 낮은 점수를 찾아서, 합계 점수 조건을 가지고 row 데이터 검색
df[df['합계'] == df[df['학년'] == 3]['합계'].min()]
```

```python
df['평균'] = df['합계']/5
```

```python
### 평균 컬럼 추가, grade 컬럼에 A:>=90 ,B:>= 80 ,C,D,F: < 60 학점 출력
def grade(score):
    if score >= 90: return 'A'
    if score >= 80: return 'B'
    if score >= 70: return 'C'
    if score >= 60: return 'D'
    else: return 'F'
```

```python
df['grade'] = df['평균'].apply(grade)
df
```

```python
# SW특기는 모두 대문자/ NaN엔 python 대문자로 넣기
# 1. 컬럼을 대문자로 변경하는 방법
# - 결측치 부여 후, str함수 사용 --> 대문자로 변경
# 2. 함수 사용, NaN 데이터 처리+대문자 변경 구문 사용
```

```python
df['SW특기'] = df['SW특기'].fillna('python').str.upper()
df
```

```python
def upper_change(txt):
    if pd.isnull(txt):
        return "python".upper()
    else: return txt.upper()
```

```python
df['SW특기'] = df['SW특기'].apply(upper_change)
df
```
