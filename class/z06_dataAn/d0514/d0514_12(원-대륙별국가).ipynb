```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 설정
matplotlib.rcParams['font.size'] = 10 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df= pd.read_csv('../z20_data/csv/drinks.csv')
df.head(3)
```

```python
df.fillna('OT',inplace=True)
```

```python
#대륙별 나라 개수 출력
# df.groupby('continent')['country'].count().sort_values(ascending=False)
df['continent'].value_counts().values
```

```python
#원그래프로 출력
labels = df['continent'].value_counts().index
values = df['continent'].value_counts().values
explode = [0.1,0,0,0,0,0]
wedgeprops = {'width':0.7,'edgecolor':'w','linewidth':2}
plt.pie(values,labels=labels,explode=explode,autopct='%.2f%%',startangle=90,counterclock=False,
        wedgeprops=wedgeprops,pctdistance=0.7)
plt.title('대륙별 국가의 수')
plt.legend(loc=(1.1,0.3),title='대륙')
```
