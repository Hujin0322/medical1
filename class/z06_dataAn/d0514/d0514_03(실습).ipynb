```python
# 이름과 국어 성적을 가지고 선 그래프 출력
import matplotlib.pyplot as plt
import matplotlib 

matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 표시

# 한글 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows 사용자
```

```python
x = list(df['이름'])
y = list(df['국어'])
plt.title('빅데이터반 성적 그래프')
plt.bar(x,y,label='국어점수')
plt.grid(axis='y',alpha=0.2,linestyle='--',linewidth=2,color='purple')
plt.legend(loc='upper center')
```

```python
# chipotle.tsv에서 가장 많이 주문한 메뉴 5개(quantity 고려O)를 꺾은 선 그래프로 출력
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
filt1 = df.groupby('item_name').sum()['quantity'].sort_values(ascending=False).head()
# df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head()
# df.groupby('item_name').sum().sort_values('quantity',ascending=False)['quantity'].head()

# 아이템별 주문한 횟수 (quantity 고려X)
filt2 = df['item_name'].value_counts()[:5]
x = filt1.index
y = filt1.values
y2 = filt2.values

plt.figure(dpi=200)
plt.plot(x,y,label='주문수')
plt.plot(x,y2,label='주문횟수')
plt.legend(loc='upper right')
```
