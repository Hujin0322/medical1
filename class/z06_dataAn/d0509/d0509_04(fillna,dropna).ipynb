```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
import numpy as np
df['학교'] = np.nan
```

#### dropna - Nan 처리: 지우기
axis: index,columns/
how: any, all

```python
# Nan 존재 컬럼만 전체 삭제
df.dropna(axis='columns',inplace=True)
```

```python
df
```

```python
# Nan 존재O 컬럼에서 모든 컬럼이 Nan인 것만 삭제
df.dropna(axis='columns',how='all',inplace=True)
```

```python
df
```

```python
# Nan O 컬럼에서 1개라도 Nan 포함 시 삭제
df.dropna(axis='columns', how='any',inplace=True)
```

```python
df
```

```python
df.loc['4번','영어'] = np.nan
```

```python
df
```

```python
df.dropna(axis='columns',how='all',inplace=True)
```

```python
df
```

```python
# Nan O index에서 1개라도 Nan 포함 시 삭제
df.dropna(axis='index', how='any',inplace=True)
df
```
