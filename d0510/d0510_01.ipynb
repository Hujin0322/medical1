```python
import pandas as pd
df = pd.read_excel('../z20_data/20122022_출생.xlsx',skiprows=2,nrows=3,usecols='B:M',index_col=0)
df.index.name = '목록'
df.iloc[0] = df.iloc[0].str.replace(",","").astype(int)

# 컬럼 추가
df['합계'] = 0

# 위치 지정 후 데이터 입력 수정.
df.iloc[0,-1] = df.iloc[0].sum()
df
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
import numpy as np
df.loc['3번','영어'] = np.nan
df.loc['4번','수학'] = np.nan
df.loc['5번',['영어','국어']] = np.nan
df
```

```python
# 총합 컬럼 만들어서, 국어,영어,수학,과학,사회의 합계 점수를 총합에 입력
df.fillna(0,inplace=True)
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = '불합격'
df.loc[df['총합'] > 300,'결과'] = '합격'

df.loc[df['총합'] > 400, '결과'] = '우수'
df
```

```python
# 총합, 결과를 사회 뒤쪽에 옮겨서 출력
cols = list(df.columns)
df = df[cols[0:8]+cols[9:]+[cols[8]]]
```
