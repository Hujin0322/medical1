```python
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
df
```

```python
# 메뉴별 주문 수량 출력
# value_counts: 컬럼별 데이터 갯수 / # sort_values: 컬럼별 데이터 정렬
df.groupby('item_name').size()
df['item_name'].value_counts()
df.groupby('item_name').count().sort_values('order_id',ascending=False)['order_id'][:10]
```

```python
df['item_name'].value_counts()[0]       # 갯수 출력
df['item_name'].value_counts().index[0] # 메뉴명 출력
```

```python
# items()

list(df['item_name'].value_counts()[:10].items())

item_count = df['item_name'].value_counts()

# items(): 항목과 갯수가 튜플 형태로 넘겨짐.
for i, (val,cnt) in enumerate(item_count.items()):
    print("Top",(i+1),":",val,",",cnt)
```

```python
# 주문 당 평균 계산 금액
df['item_price'] = df['item_price'].str.replace('$','').astype(float)
df.groupby('order_id')['item_price'].sum().mean()
```

```python
# item 당 주문 갯수 출력
order_count = df.groupby('item_name')['order_id'].count()
order_count[:10]
```

```python
# item 당 주문 총량 출력
len(df.groupby('item_name'))
df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head()
```

```python
# 한 주문에 10달러 이상 사용한 주문의 id 출력
order_id_group = df.groupby('order_id').sum()
df.groupby('order_id').sum()['item_price'] > 20
# order_id_group[order_id_group['item_price']>20]

order_id_group.loc[df.groupby('order_id').sum()['item_price'] > 20,['item_price']].head()
```

```python
# 각 아이템별 가격을 구하시오.
# 1. 아이템별 주문 갯수 = 1의 가격 출력
quan = df[df['quantity']==1]
# 2. 50개별 최저 격 = 아이템 별 1개 주문 금액
price = quan.groupby('item_name').min().sort_values('item_price',ascending=False)
price[['item_price']].head()
# 메뉴 갯수: 50
# len(quan.groupby('item_name').min().sort_values('item_price')['item_price'])
```

```python
# 주문 당 갯수가 가장 많은 상위 5개 출력
df.groupby('order_id').sum().sort_values('quantity',ascending=False)[['quantity','item_price']].head()
```

```python
# Veggie Salad Bowl 메뉴가 몇 번 주문 되었는지 출력
len(df[df['item_name'] == 'Veggie Salad Bowl'])
# len(df[df['item_name'] == 'Veggie Salad Bowl']['order_id'].unique())
df[df['item_name'] == 'Veggie Salad Bowl'].sort_values('item_price',ascending=False)
```
